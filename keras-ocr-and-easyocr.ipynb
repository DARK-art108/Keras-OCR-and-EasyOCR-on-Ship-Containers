## IMPORTS

```python
import pytesseract
import keras_ocr
import easyocr
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

## TEST IMAGES

```python
url = [
    "https://5.imimg.com/data5/UK/CH/FV/SELLER-11338434/40-hc-used-shipping-container-in-mundra-500x500.jpg",
    "https://cdn.shopify.com/s/files/1/0120/4849/8752/products/20STCW_737x.jpg?v=1579271736",
]
images = [ keras_ocr.tools.read(i) for i in url]
```

```python
fig = plt.figure(figsize=(16,10))
rows = 1
columns = 1

fig.add_subplot(rows, columns, 1)
plt.imshow(images[0])
plt.axis('off')
plt.title("First Image")
```

```python
fig = plt.figure(figsize=(16,10))
rows = 1
columns = 1

fig.add_subplot(rows, columns, 1)
plt.imshow(images[1])
plt.axis('off')
plt.title("Second Image")
```

<a id="kerasocr"></a>
# KERAS_OCR

```python
pipline = keras_ocr.pipeline.Pipeline() #Creting a pipline 
kerasocr_preds = pipline.recognize(images)
```

## Results of Keras-OCR

```python
fig,axs = plt.subplots(nrows = len(images) , figsize = (30,30))
for ax , image,  prediction in zip(axs , images , kerasocr_preds):
    keras_ocr.tools.drawAnnotations(image, prediction, ax)
```

<a id="easyocr"></a>
# EASYOCR

```python
text_reader = easyocr.Reader(['en']) #Initialzing the ocr
```

## Results of EASY OCR

```python
results = text_reader.readtext(images[0] )
for (bbox, text, prob) in results:
    print(text)
plt.imshow(images[0])
plt.title("First Image");
```

```python
results = text_reader.readtext(images[0] )
for (bbox, text, prob) in results:
    print(text)
plt.imshow(images[1])
plt.title("Second Image");
```
